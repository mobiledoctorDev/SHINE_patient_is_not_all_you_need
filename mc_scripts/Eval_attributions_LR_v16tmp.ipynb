```python
import pickle
from sklearn.linear_model import LogisticRegression
import pandas as pd

pd.set_option('display.max_columns', 120)
```

```python
def get_attr_result(save_dir):
    df_attr = pd.DataFrame()
    for i in range(5):
        filename = f"{save_dir}/model_LR{i}.pkl"
        loaded_model = pickle.load(open(filename, 'rb'))
        print(f"{i} n_iter_:", loaded_model.n_iter_)
        model_name = "+".join(save_dir.split("_")[4:5])
        df_1 = pd.DataFrame([loaded_model.feature_names_in_, loaded_model.coef_[0]])
        df_1.columns = df_1.iloc[0]
        df_1['feature'] = model_name
        df_1['intercept'] = loaded_model.intercept_[0]
        df_1['n_feature'] = loaded_model.n_features_in_
        df_1['i'] = i
        df_1 = df_1.loc[1:, :]
        head_feature = ['feature', 'n_feature', 'i' , 'intercept']
        df_1 = df_1[head_feature + [x for x in df_1.columns if x not in head_feature]]
        
        if df_attr.empty:
            df_attr = df_1.copy()
        else:
            df_attr = pd.concat([df_attr, df_1])

    df_attr = df_attr.reset_index(drop=True)
    return df_attr

df_patient = get_attr_result('./model/train_result/v16tmp_20230706_235307_patient_manual_fillna')
df_pg = get_attr_result('./model/train_result/v16tmp_20230706_235310_patient+gps_manual_fillna')
df_ps = get_attr_result('./model/train_result/v16tmp_20230706_235313_patient+si10_manual_fillna')
df_po = get_attr_result('./model/train_result/v16tmp_20230706_235315_patient+owid_manual_fillna/')
df_all = get_attr_result('./model/train_result/v16tmp_20230706_235319_patient+gps+si10+owid_manual_fillna/')
```

```python
df_concat = pd.concat([df_patient, df_pg, df_ps, df_po, df_all])
```
