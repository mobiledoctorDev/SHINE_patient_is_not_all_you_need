# Save by month

```python
new_target_months_list = []
for yyyy_m in target_months_list:
    yyyy, m = yyyy_m.split("_")
    yyyy_mm = "{:04d}_{:02d}".format(int(yyyy), int(m))
    new_target_months_list.append(yyyy_mm)
new_target_months_list
```

```python
for idx, yyyy_mm in enumerate(new_target_months_list):
    
    start_yyyy_mm = yyyy_mm.replace("_", "-") + "-01 00:00:00"
    if idx == len(new_target_months_list) - 1:
        next_yyyy_mm = "2021-06-01 00:00:00"
    else:
        next_yyyy_mm = new_target_months_list[idx+1].replace("_", "-") + "-01 00:00:00"
    print(yyyy_mm, start_yyyy_mm, next_yyyy_mm)
    target_cols = df_all_md_data.columns.tolist() + ["birthday", 'convulsion']
    df_tmp = df_join[(df_join['date'] >= start_yyyy_mm ) & (df_join['date'] < next_yyyy_mm )][target_cols]
    print(yyyy_mm, df_tmp.shape)
    
    filepath = "/Users/mckim/PycharmProjects/Apollo_model_dev/data/v2/md_data_{}.csv".format(yyyy_mm)
    print(filepath)
    df_tmp.to_csv(filepath, index=False)
    
```
